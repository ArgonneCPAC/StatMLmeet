# Script to read the data

Datasets are stored in 2 HDF5 files, for z=50 and z=0 respectively. Both files have 100 realizations. 

Requires h5py (conda install h5py)


```python
import h5py
import matplotlib.pylab as plt
import numpy as np


simID = 32 # nTotal = 100
slice = 12 # nGrids = 64

z = 50
hdf5_path = "../data/z" + str(z)+"_100.hdf5"
hdf5_file = h5py.File(hdf5_path, mode='r')
sim_z50 = hdf5_file["sims_z" + str(z)]

z = 0
hdf5_path = "../data/z" + str(z)+"_100.hdf5"
hdf5_file = h5py.File(hdf5_path, mode='r')
sim_z0 = hdf5_file["sims_z" + str(z)]
```

```python
### Plotting slices of a 
sim_z0 = np.log10(sim_z0)
```

```python
fig, ax = plt.subplots(2, 3, figsize = (16,  10))
ax[0, 0].imshow(sim_z50[simID, slice, :, :], cmap='gray_r')
ax[0, 1].imshow(sim_z50[simID, :, slice, :], cmap='gray_r')
ax[0, 2].imshow(sim_z50[simID, :, :, slice], cmap='gray_r')

ax[1, 0].imshow(sim_z0[simID, slice, :, :], cmap='gray_r')
ax[1, 1].imshow(sim_z0[simID, :, slice, :], cmap='gray_r')
ax[1, 2].imshow(sim_z0[simID, :, :, slice], cmap='gray_r')

ax[1, 0].set_xlabel('X ($h^{-1}$ Mpc)',fontsize=30)
ax[1, 1].set_xlabel('Y ($h^{-1}$ Mpc)', fontsize=30)
ax[1, 2].set_xlabel('Z ($h^{-1}$ Mpc)', fontsize=30)

ax[0, 0].set_ylabel('z = 50', rotation=0, fontsize=30, labelpad=50)
ax[1, 0].set_ylabel('z = 0', rotation=0, fontsize=30, labelpad=50)


plt.savefig('zeldo_snaps.pdf')
plt.show()
```
